```python
import pandas as pd
url="https://raw.githubusercontent.com/ourcodingclub/CC-python-pandas-matplotlib/master/scottish_hills.csv"
df=pd.read_csv(url)
```

```python
print(df.head(10))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x=df.Height
y=df.Latitude
z=df.Longitude
plt.scatter(x,y)
plt.savefig("scottish_scatter_plot.png")
```

```python
from scipy.stats import linregress
stats=linregress(x,y)
m=stats.slope
b=stats.intercept
plt.scatter(x,y)
plt.plot(x,m*x+b,color="red")
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y,marker="x")
plt.plot(x,m*x+b,color="red",linewidth=3)
plt.xlabel("Height(m)",fontsize=20)
plt.ylabel("Latitude",fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
```

```python
import numpy as np
colors=np.random.rand(len(y))
plt.scatter(y,x,s=(x-900),c=colors,alpha=0.5)
```

```python
plt.hist(x,bins=25,density=True)
plt.savefig("his.png",dpi=25)
```

```python
def f_c(temp):
    return (5./9.)*(temp-32)
temp=[100,102,106,105,90,85,85,89,100,102,103,108,100,102,106,105,90,85,85,89,100,102,103,108]
```

```python
fig,ax_f=plt.subplots()
ax_c=ax_f.twinx()
ax_f.plot(temp)
ax_c.set_xlim(0,24)
y1,y2=ax_f.get_ylim()
ax_c.set_ylim(f_c(y1),f_c(y2))
ax_c.figure.canvas.draw()
ax_f.set_title("2 scales:F C")
ax_f.set_ylabel("F")
ax_c.set_ylabel("c")
ax_f.set_xlabel("t")

```

```python
new_url="https://raw.githubusercontent.com/plotly/datasets/master/school_earnings.csv"
df1=pd.read_csv(new_url)
print(df1.head(10))
```

```python
#his
x1=df1.Women
y1=df1.Men
z1=df1.Gap
plt.hist(x1,bins=25,density=True)
plt.savefig("his1.png",dpi=25)
```

```python
plt.style.use("seaborn-pastel")
fig,ax=plt.subplots()
ax = plt.gca()
ax.set_facecolor('xkcd:salmon')
ax.set_facecolor((1.0, 0.47, 0.42))
ax.hist(x1,bins=25,density=True,histtype="stepfilled",alpha=0.8,label="women")
ax.hist(y1,bins=25,density=True,histtype="stepfilled",alpha=0.8,label="men")
ax.legend(prop={'size':10})
ax.set_ylabel("f")
ax.set_xlabel("salaary")
```

```python
labels=z1
index=np.arange(len(labels))
fig,ax=plt.subplots()
bar_width=0.35
opacity=0.6
rects1=ax.bar(index,x1,bar_width,alpha=opacity,color="b",label="women")
rects2=ax.bar(index+bar_width,y1,bar_width,alpha=opacity,color="r",label="men")
ax.set_xlabel("school")
ax.set_ylabel("salary")
ax.set_title("salray for both gender per school")
ax.set_xticks(index+bar_width/2)
ax.set_xticklabels(df1.School, rotation='vertical')######rotate
#plt.xticks(x, labels, rotation='vertical')
ax.legend()
```
