```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA


faces = fetch_lfw_people(min_faces_per_person=60)

pca = PCA(150).fit(faces.data)#150*original dimension 
components = pca.transform(faces.data)
projected = pca.inverse_transform(components)

fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(faces.data[i].reshape(62, 47), cmap='binary_r')
    ax[1, i].imshow(projected[i].reshape(62, 47), cmap='binary_r')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('150-dim\nreconstruction');

plt.show()


fig, axes = plt.subplots(3, 8, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(62, 47), cmap='bone')

plt.show()
```

```python
#PCA(principal component analysis)
#clustering data with data dimension reduction

#1)))).find the best "line" that represent the distribution of data 
#(minimize the distance of that points to the "line"=====max the distance of the projection of the points onto the lines to the origin)

#sum of the squared distance of projected point to origin===SS(distances)
#max is the best line---PC1
#(slope gradient tells about how important certain variable contributes to the data distribution)--linear combination of variables

#the unit vector of vector of this line ======"eigenvector" of PC1

#SS(distances)=====eigenvalues for pc1
#sqrt(eigenvalues)=singular value for PC1

#PC2 is perpendicular to PC1

#2}}} then pc1 and pc2 become two axis. the points projection on the 2 axis become the new representation of these points


#3))))) varaince:    variation for pc1===SS(distances for pc1)/(n-1);   variation for pc2===SS(distances for pc2)/(n-1); 

#the porportion of eigenvalues of each pc credit the importance of this pc for the variance of data.
#hence: dim reduction just use high eigenvalue set of pcs.(if eigenvalues are not so high then, reduction ineffective!)

```
